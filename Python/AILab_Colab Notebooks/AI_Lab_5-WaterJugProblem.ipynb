```python
def water_jug(x,y,target):
  a,b=0,0

  while True:
    print(a,b)

    if a==target or b==target:
      print("Reached the Target : ")
      break



    if a==0:
      a = x

    elif b==y:
      b = 0

    else:
      pour = min(a,y-b)
      a -= pour
      b += pour

```

```python
water_jug(2,3,1)
water_jug
```
